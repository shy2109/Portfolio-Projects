```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (14,8)

pd.options.mode.chained_assignment = None

#Loading the data
df = pd.read_csv(r'C:\Users\ULTIMATE\Downloads\Spotify Dataset\tracks.csv')

df.head()
```

```python
#checking null values in the dataset
pd.isnull(df).sum()
```

```python
df.info()
```

```python
sorted_df = df.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
df.describe().transpose()
```

```python
#top 10 most popular songs on Spotify
most_popular = df.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
#changing the index of the dateframe to release date
df.set_index('release_date', inplace = True)
df.index= pd.to_datetime(df.index)
df.head()
```

```python
#converting the duration to seconds
df["duration"] = df["duration_ms"].apply(lambda x: round(x/1000))
df.drop("duration_ms", inplace = True, axis = 1)
df.duration.head()
```

```python
#creating pearson correlation visualization
corr_df= df.drop(["key", "mode", "explicit"], axis =1).corr(method= "pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidths= 1, linecolor= "Black")
heatmap.set_title("Correlation Heatmap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
#sampling the data to create regression plots
sample_df = df.sample(int(0.004*len(df)))
len(sample_df)
```

```python
#visualizing the positive correlation between loudness and energy
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y= "loudness", x= "energy", color = 'c').set(title = "Loudness vs Energy Correlation")
```

```python
#visualizing the positive correlation between popularity and acousticness
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y= "popularity", x= "acousticness", color = 'r').set(title = "Popularity vs Acousticness Correlation")
```

```python
#creating new column "year" extracted from release date
df['dates'] = df.index.get_level_values('release_date')
df.dates = pd.to_datetime(df.dates)
years = df.dates.dt.year
```

```python
#Plotting number of songs released per year
sns.distplot( years).set(title = "Number of songs per year")
```

```python
#duration of songs every year
total_dur = df.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize= fig_dims)
fig = sns.barplot(x = years, y = total_dur, ax = ax,errwidth= False).set(title = "Years vs Duration")
plt.xticks(rotation = 90)
```

```python
#analysing average duration of songs in a lineplot

total_dur = df.duration
fig_dims = (10,5)
fig, ax = plt.subplots(figsize= fig_dims)
fig = sns.lineplot(x = years, y = total_dur, ax = ax).set(title = "Years vs Duration")
plt.xticks(rotation = 90)
```
